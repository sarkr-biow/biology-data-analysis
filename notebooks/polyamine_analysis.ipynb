# Biological Data Analysis: Polyamine Pathways

```python
# Import core libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: configure plots
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Simulate a dataset for polyamine pathway genes
data = {
    "Gene": ["speA", "speB", "speC", "speD", "adiA"],
    "Expression_Control": [12.5, 8.1, 15.0, 6.2, 5.5],
    "Expression_Treatment": [18.2, 4.5, 20.1, 3.9, 6.0]
}

df = pd.DataFrame(data)
df
```

```python
# Compute fold change
df["Fold_Change"] = df["Expression_Treatment"] / df["Expression_Control"]
df
```

```python
# Bar plot of gene expression
plt.figure(figsize=(8,5))
sns.barplot(x="Gene", y="Expression_Treatment", data=df, color="skyblue", label="Treatment")
sns.barplot(x="Gene", y="Expression_Control", data=df, color="lightgreen", label="Control")
plt.ylabel("Expression Level")
plt.title("Gene Expression: Control vs Treatment")
plt.legend()
plt.show()
```
